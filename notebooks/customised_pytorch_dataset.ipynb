```python
## Convert data into a dataset
list_mz_0 = [m.spectrum_object_0.mz_array for m in molecule_pair]
list_intensity_0 = [m.spectrum_object_0.intensity_array for m in molecule_pair]
list_mz_1 = [m.spectrum_object_1.mz_array for m in molecule_pair]
list_intensity_1 = [m.spectrum_object_1.intensity_array for m in molecule_pair] 
list_similarity = [m.similarity for m in molecule_pair]
```

```python
mz_0 =np.zeros((len(molecule_pair),500))
intensity_0 = np.zeros((len(molecule_pair),500))
mz_1=np.zeros((len(molecule_pair),500))
intensity_1 = np.zeros((len(molecule_pair),500))
similarity= np.zeros((len(molecule_pair),1))
```

```python
import torch
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    def __init__(self, your_dict):
        self.data = your_dict
        self.keys = list(your_dict.keys())

    def __len__(self):
        return len(self.data[self.keys[0]])
        #return len(self.keys)

    def __getitem__(self, idx):
        #key = self.keys[idx]
        #sample = self.data[key]
        #print(idx)
        sample = {k:self.data[k][idx] for k in self.keys}
        # Convert your sample to PyTorch tensors if needed
        # e.g., use torch.tensor(sample) if sample is a numpy array
        
        return sample
```

```python
dataset = CustomDataset(dictionary_data)
```

```python
# Creating a DataLoader (optional but useful for batching)
from torch.utils.data import DataLoader

dataloader = DataLoader(dataset, batch_size=1, shuffle=True)

```

```python
# Accessing batches in the dataloader
for batch in dataloader:
    sample = batch
    #print(f'Batch Keys: {keys}, Batch Samples: {samples}')
```
