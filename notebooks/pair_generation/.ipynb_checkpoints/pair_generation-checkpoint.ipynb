## STEPS:
## 1) Get the standard smiles 
## 2) Get a table with standard smiles and corresponding indexes in terms of spectrums

## 4) Run pair generation on it

* get the standard smiles

```python
from rdkit import Chem
import numpy as np
import pandas as pd
```

```python

canon_smiles= [Chem.CanonSmiles(s.smiles) for s in all_spectrums_train]
```

```python
len(canon_smiles)
```

```python
unique_smiles = np.unique(canon_smiles)
```

```python
len(unique_smiles)
```

```python
#indexes 
table_smiles = {s: [i for i,c in enumerate(canon_smiles) if c==s] for s in unique_smiles}
```

```python
all_mz= [s.precursor_mz for s in all_spectrums_train]
```

```python
df_smiles = pd.DataFrame()
df_smiles['canon_smiles']=list(unique_smiles)
df_smiles['indexes'] = [table_smiles[k] for k in unique_smiles]
df_smiles['mz'] =    [all_mz[canon_smiles.index(u_s)] for u_s in unique_smiles]
```
